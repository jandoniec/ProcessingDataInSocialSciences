The files krk-airlines.csv, krk-flights.csv and krk-passengers.csv contain data about flights from Krakow Airport. In a separate notebook, calculate and display:

- list including flight number and destination (two columns)
- flight list with the full name of the airline and the name of the aircraft
- a list of passengers on a flight to London sorted by surname
- a list of women flying to Paris ordered by surname
- number of passengers on the flight to Berlin
- number of men flying from Krakow
- number of passengers for each flight

```python
import pandas as pd
```

```python
krk_airlines = pd.read_csv('krk-airlines.csv')
krk_flights = pd.read_csv('krk-flights.csv')
krk_passengers = pd.read_csv('krk-passengers.csv')
```

```python
krk_flights.head()
```

```python
krk_airlines.head()
```

```python
krk_passengers.head()
```

```python
krk_flights[['flight', 'to']]
```

```python
krk_flights.merge(krk_airlines, on='airlineid')[['flight', 'airline', 'airplane']]
```

```python
krk_passengers.merge(krk_flights, on='flight').query('to == "London"').sort_values('surname')[['name', 'surname']]
```

```python
krk_passengers.merge(krk_flights, on='flight').query('to == "Paris" and gender == "Female"').sort_values('surname')[['name', 'surname']]
```

```python
krk_passengers.merge(krk_flights, on='flight').query('to == "Berlin"').shape[0]
```

```python
krk_passengers.merge(krk_flights, on='flight').query('gender == "Male"').shape[0]
```

```python
krk_flights.merge(krk_passengers, on='flight').groupby('flight').count()[['name']]
```
