Calculate and display the total number of students studying in Warsaw and Krakow. The data is available in the universitites.csv file. Do your calculations in a separate notebook.

```python
import pandas as pd
```

```python
srudenci=pd.read_csv('universities.csv')
srudenci
```

```python
srudenci['Students'] = pd.to_numeric(srudenci['Students'])
total_students = srudenci.loc[srudenci['City'].isin(['Warsaw', 'Krakow']), 'Students'].sum()
total_students
```

```python
krk = srudenci.loc[srudenci['City'].isin(['Krakow']), 'Students'].sum()
krk
```

```python
wwa = srudenci.loc[srudenci['City'].isin(['Warsaw']), 'Students'].sum()
wwa
```
