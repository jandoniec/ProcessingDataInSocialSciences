The employees.csv file contains data about employees. Using Power Query, import the data into a spreadsheet. Then:

display the sorted the data by salary, starting with the highest
display employees from Poland, aged at least 50 years
calculate the number of women younger than 50

```python
import pandas as pd
```

```python
employees = pd.read_csv('employees.csv')
employees.sort_values('salary', ascending=False)
```

```python
employees.loc[(employees['country'] == 'Poland') & (employees['age'] >= 50)]
```

```python
count = len(employees.loc[(employees['sex'] == 'Female') & (employees['age'] < 50)])
count
```
